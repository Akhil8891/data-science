```python
import matplotlib.pyplot as plt
import numpy as np

plt.suptitle("Graph")

# Data for Subplot 1
x = np.array([0, 1])
y = np.array([2, 6])

# Subplot 1
plt.subplot(2, 3, 1)
plt.title("Title 1")
plt.xlabel("X-label", color="b")
plt.ylabel("Y-label", color="b")
plt.plot(x, y, 'o', color='blue')

# Data for Subplot 2
y1 = np.array([3, 8, 1, 10])
y2 = np.array([8, 4, 7, 3])

# Subplot 2
plt.subplot(2, 3, 2)
plt.plot(y1, marker="*", color="orange", lw=2.5)
plt.plot(y2, linestyle="dotted", color="blue", lw=2.5)

# Data for Subplot 3
y = np.array([4, 8, 5, 10])

# Subplot 3
plt.subplot(2, 3, 3)
plt.plot(y, marker="o", mfc="y", mec="k")

# Data for Subplot 4
x = np.array([1, 8])
y = np.array([3, 10])

# Subplot 4
plt.subplot(2, 3, 4)
plt.title("Title 2")
plt.plot(x, y)
plt.grid()

# Data for Subplot 5 (Scatter Plot)
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([10, 20, 15, 25, 66, 33, 88, 22, 14, 30, 4, 57, 60])
colors = np.array(["red", "green", "blue", "yellow", "pink", "black", "orange", "purple", "beige", "brown", "gray", "cyan", "magenta"])
sizes = np.array([50, 80, 100, 60, 90, 120, 200, 70, 65, 150, 45, 110, 130])

# Subplot 5
plt.subplot(2, 3, 5)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)

# Data for Subplot 6 (Bar Plot)
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 8, 9])

# Subplot 6
plt.subplot(2, 3, 6)
plt.bar(x, y, color="r")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["red", "blue", "yellow", "green"]

# Create the pie chart
plt.pie(y, labels=mylabels, startangle=90, explode=myexplode, shadow=True, colors=mycolors)

# Add a legend manually (you can customize this part as needed)
plt.legend(mylabels, title="Fruits")

# Show the plot
plt.show()
```
